# 2.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello, World!"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

# 3.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello, World!"
   
@app.route("/welcome")
def hello_world1():
    return "Welcome to ABC corporation"

    
if __name__=="__main__":
    app.run(host="0.0.0.0")
```
