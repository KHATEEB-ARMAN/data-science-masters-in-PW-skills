## 2.

```python
import seaborn as sns
```

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal)
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal,hue=fmri.event)
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal,size=fmri.event,hue=fmri.region)
```
