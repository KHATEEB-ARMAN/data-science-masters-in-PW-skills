```python
# K-means has difficulty with non-globular clusters and clusters of multiple sizes. DBSCAN is used to handle clusters of multiple
# sizes and structures and is not powerfully influenced by noise or outliers. K-means can be used for data that has a clear centroid,
# including a mean or median.
```
